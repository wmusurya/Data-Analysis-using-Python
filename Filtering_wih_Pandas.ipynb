```python
d={'Make':['Toyota','Honda','Toyota','BMW','Nissan','Toyota','Honda','Honda','Toyota','Nissan'],'Color':['White','Red','Blue','Black','White','Green','Blue','Blue','White','White'],'Odometer':[150043,87899,32549,11179,213095,99213,45698,54738,60000,31600],
  'Doors':[4,4,3,5,4,4,4,4,4,4],'Price':[4000,5000,7000,22000,3500,4500,7500,7000,62500,9700]}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(d)
```

```python
df
```

```python
Question: What is the average price for all cars
Question: How many cars have 4 doors?
Question: What is the total odometer reading for Nissan cars?
Question: Which car has the highest price, and what is the price?
Question: What is the average odometer reading for Toyota cars?
Question: How many cars are there for each make in the DataFrame?
Answer:
Toyota: 4 cars
Honda: 3 cars
BMW: 1 car
Nissan: 2 cars
Question: What is the lowest price among the Honda cars?
Question: What is the total number of doors for all cars
Question: What is the most common color among the cars
Question: What is the average price for Honda cars?
```

```python
#Question: What is the average price for all cars
df[['Price']].mean()
```

```python
#How many cars have 4 doors
a=df.query('Doors==4')
a
```

```python
a[['Make','Doors']]
```

```python
df[df['Doors']==4]['Make'].unique()
```

```python
df
```

```python
#What is the total odometer reading for Nissan cars?

df[df['Make']=='Nissan']['Odometer'].sum()
```

```python
#Which car has the highest price, and what is the price?

df[df['Price']==df['Price'].max()][['Make','Price']]
```

```python
#What is the average odometer reading for Toyota cars?

df[df['Make']=='Toyota']['Odometer'].mean()
```
